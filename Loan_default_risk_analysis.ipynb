**Data Profiling**

```python
df.describe().T # T for transpose
```

```python
# Select only numerical columns
num_cols = df.select_dtypes(include='number')
# num_cols dA variable shayl kol table el loan wa (select_dtypes) bt7dad no3 mo3yn mn elcolumns ya get included fy num columns zy hena numbers bs


# Create profiling summary
profiling = pd.DataFrame({
    "Count": num_cols.count(),
    "Missing Values": num_cols.isnull().sum(),
    "Mean": num_cols.mean(),
    "Median (Q2)": num_cols.median(),
    "Min": num_cols.min(),
    "Q1": num_cols.quantile(0.25),
    "Q3": num_cols.quantile(0.75),
    "Max": num_cols.max()
})

# Calculate IQR
profiling["IQR"] = profiling["Q3"] - profiling["Q1"]

# Calculate Outlier Limits
profiling["Lower Limit"] = profiling["Q1"] - 1.5 * profiling["IQR"]
profiling["Upper Limit"] = profiling["Q3"] + 1.5 * profiling["IQR"]

# tb3n kol da 3shan acheck fy outliers wla la 3shan bashof fy haga msln  ta7t el loweer limit wla la
# Round numbers
profiling = profiling.round(2)

profiling
```
